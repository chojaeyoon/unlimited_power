```
# test dataset 불러오기
# 대회에서는 불러오면 실격, 잘 학습이 되었나 확인만 하기 위함

test_dir = '/content/drive/MyDrive/Jupyter/unlimited_power/raw_data/test'
test_list = os.listdir(test_dir)
test_list[:10]
```

```
# 순서가 맞지 않는다. 순서를 맞춰야 한다.
# 순서를 맞추는 함수

import re

_nsre = re.compile('([0-9]+)')
def natural_sort_key(s):
    return [int(text) if text.isdigit() else text.lower()
            for text in re.split(_nsre, s)]  
```

```
# 만족스럽다.

test_list.sort(key = natural_sort_key)
test_list[:10]
```

```
# 실제로 자료 불러오기

test_data = [[] for i in range(0,len(test_list))]

for i in range(0, len(test_list)):
    file_name = test_dir + '/' + test_list[i]
    test_data[i] = pd.read_csv(file_name, delimiter=',')
    test_data[i].drop('Day', axis = 1, inplace = True)

# 불러온 친구들 합치기
test_dataframe = pd.DataFrame()
for i in range(0, len(test_list)):
    test_dataframe = concat([test_dataframe, test_data[i]], axis = 0)

test_dataframe.reset_index(inplace=True)
test_dataframe.drop('index', axis = 1, inplace = True)

# 완성본

test_dataframe
```
