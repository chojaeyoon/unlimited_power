```python
raw_data = pd.read_csv('/content/drive/MyDrive/Jupyter/unlimitied_power/raw_data/train/train.csv')
```

```python
# test dataset 불러오기
# 대회에서는 불러오면 실격, 잘 학습이 되었나 확인만 하기 위함

test_dir = '/content/drive/MyDrive/Jupyter/unlimitied_power/raw_data/test'
test_list = os.listdir(test_dir)
test_list[:10]
```

```python
# 순서가 맞지 않는다. 순서를 맞춰야 한다.
# 순서를 맞추는 함수

import re

_nsre = re.compile('([0-9]+)')
def natural_sort_key(s):
    return [int(text) if text.isdigit() else text.lower()
            for text in re.split(_nsre, s)]  
```

```python
# 만족스럽다.

test_list.sort(key = natural_sort_key)
test_list[:10]
```

```python
# 실제로 자료 불러오기

test_data = [[] for i in range(0,len(test_list))]

for i in range(0, len(test_list)):
    file_name = test_dir + '/' + test_list[i]
    test_data[i] = pd.read_csv(file_name, delimiter=',')
    test_data[i].drop('Day', axis = 1, inplace = True)

# 불러온 친구들 합치기
test_dataframe = pd.DataFrame()
for i in range(0, len(test_list)):
    test_dataframe = concat([test_dataframe, test_data[i]], axis = 0)

test_dataframe.reset_index(inplace=True)
test_dataframe.drop('index', axis = 1, inplace = True)

# 완성본

test_dataframe
```

```python
# LSTM에 넣기 전에 스케일링

scaler = MinMaxScaler()
scaled_train = pd.DataFrame(scaler.fit_transform(raw_data), columns=raw_data.columns, index = raw_data.index)
scaled_test = pd.DataFrame(scaler.fit_transform(test_dataframe), columns=test_dataframe.columns, index = test_dataframe.index)
scaled_train
```

```python
# Day는 상관 없는 자료이니, 날리고, 시간(hour, minute)은 시간과 일조량의 상관관계가 있을테니 혹시 몰라 남겨본다.
scaled_train = scaled_train.drop('Day', axis = 1)
```

```python
# 과거 7일의 데이터를 가지고 예측
# 하루에 48개의 tick이 생성되므로(30분 간격) 7일은 총 48*7 = 336틱
# 336 틱 데이터를 가지고 예측한다.

n_days = 336

# 사용할 특성의 개수는 컬럼의 개수와 동일
n_features = len(scaled_train.columns)

# 미래 2일을 예측
# 2일은 48 * 2 = 96틱
future_window = 96
```

```python
# 데이터 셋을 만들어주는 함수
def series_to_supervised(data, n_in=1, n_out=1, dropnan=True):
    n_vars = 1 if type(data) is list else data.shape[1]
    df = DataFrame(data)
    cols, names = list(), list()
    # input sequence (t-n, ... t-1)
    for i in range(n_in, 0, -1):
        cols.append(df.shift(i))
        names += [('var%d(t-%d)' % (j+1, i)) for j in range(n_vars)]
#    forecast sequence (t, t+1, ... t+n)
    for i in range(0, n_out):
        cols.append(df.shift(-i))
        if i == 0:
            names += [('var%d(t)' % (j+1)) for j in range(n_vars)]
        else:
            names += [('var%d(t+%d)' % (j+1, i)) for j in range(n_vars)]
    # put it all together
    agg = concat(cols, axis=1)
    agg.columns = names
    # drop rows with NaN values
    if dropnan:
        agg.dropna(inplace=True)
    return agg
```

```python
reframed_train = series_to_supervised(scaled_train, n_days, 1)
reframed_test = series_to_supervised(scaled_test, n_days, 1)
reframed_train
```

```python
train = reframed_train.values
test = reframed_test.values

# 학습할 데이터의 갯수는 틱의 갯수 * 특징의 갯수 ( 336틱 * 변수 8개)
n_obs = n_days * n_features

# 336틱 * 변수 8개는 학습 데이터 train_X로 분리
# 마지막의 마지막에 var8(t)가 우리가 원하는 그날의 발전량 이므로 target은 train_y로 분리
train_X, train_y = train[:, :n_obs], train[:, -1]
test_X, test_y = test[:, :n_obs], test[:, -1]

print(train_X.shape, len(train_X), train_y.shape)

train_X = train_X.reshape((train_X.shape[0], n_days, n_features))
test_X = test_X.reshape((test_X.shape[0], n_days, n_features))
```

```python
model = Sequential()
model.add(LSTM(50, input_shape=(train_X.shape[1], train_X.shape[2])))

model.add(Dense(future_window))
model.compile(loss='mse', optimizer='adam')
```

```python
history = model.fit(train_X, train_y, epochs=100, batch_size=250, verbose=2, shuffle=False, validation_data=(test_X, test_y))
```

```python
pyplot.plot(history.history['loss'], label='train')
pyplot.plot(history.history['val_loss'], label='test')
pyplot.legend()
pyplot.show()
```
